### Dealing with missing values

```python
import missingno as msno

msno.bar(df, color="lightblue")
plt.show()
```

### Apply( ) functions

```python
def split_value(value):
    return value.split(",")
```

```python
def divide(number):
    return number / 2
```

```python
df.Genre.apply(split_value)
```

```python
df.Rating.apply(divide)
```

```python
df[["Rating", "Votes", "Runtime (Minutes)", "Revenue (Millions)", "Metascore"]].apply(
    divide
)
```

```python
df.corr()
```

```python
plt.figure(figsize=(12, 12))
sns.heatmap(df.corr(), annot=True, fmt=".3f", linewidths=2, square=True, cmap="Blues")
plt.title("Correlation", size=15)
plt.show()
```
